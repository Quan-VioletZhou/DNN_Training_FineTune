```python
# Make some scene noise:
import os
from pydub import AudioSegment

# Define directories
input_folder = r"D:\DNN\Training_data\Noise\scenes_stereo\scenes_stereo"
output_folder = r"D:\DNN\Training_data\Noise\scenes_stereo\Extracted_scenes_noise"

# Ensure output directory exists
os.makedirs(output_folder, exist_ok=True)

# Desired clip duration (2 seconds = 2000 milliseconds)
clip_duration_ms = 2000  

# Loop through all .wav files in the input folder
for file in os.listdir(input_folder):
    if file.endswith(".wav"):
        file_path = os.path.join(input_folder, file)
        noise = AudioSegment.from_wav(file_path)  # Load audio
        
        # Extract 2s clips from the 30s file
        num_clips = len(noise) // clip_duration_ms  # Should be 15 clips per file
        for i in range(num_clips):
            start = i * clip_duration_ms  # Start time
            end = start + clip_duration_ms  # End time
            clip = noise[start:end]  # Slice the audio
            
            # Save the extracted clip with a new name
            output_filename = f"{file[:-4]}_clip_{i+1}.wav"
            output_path = os.path.join(output_folder, output_filename)
            clip.export(output_path, format="wav")

            print(f"Saved: {output_filename}")

print("✅ All scene noise files have been split into 2-second clips!")
```
